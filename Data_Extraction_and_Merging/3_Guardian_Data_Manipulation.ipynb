```python
# Define keywords to filter titles
keywords = ["Refugees", "Migration", "Immigration", "Refugee", "Asylum Seeker", "Asylum Seekers", "Migrants", "Migrant", "Immigrants", "Immigrant"]

# Create a regex pattern that matches any of the keywords
pattern = '|'.join(keywords)

# Filter rows where 'webTitle' contains any of the keywords (case-insensitive)
filtered_df = df[df['webTitle'].str.contains(pattern, case=False, na=False)]

# Display or save the filtered DataFrame
#filtered_df.to_csv("filtered_file.csv", index=False)  # Save to a new CSV if needed

filtered_df.head()
```
