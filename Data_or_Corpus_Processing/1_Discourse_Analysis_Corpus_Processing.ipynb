```python
#Saving as individual text files. 

import pandas as pd
import os

# Load the CSV file
df = pd.read_csv("/Users/henryasiamah/Desktop/work/discourse_analysis_corpus.csv")

# Create a folder to store the text files (optional)
output_folder = "/Users/henryasiamah/Desktop/work/Discourse"
os.makedirs(output_folder, exist_ok=True)

# Initialize a counter for the number of files created
file_count = 0

# Iterate through each row and create a .txt file
for index, row in df.iterrows():
    # Generate the filename as "Date_Source_Index.txt" to ensure uniqueness
    date_str = row['Date'].replace("/", "-")  # Replace slashes if any to avoid issues in filename
    source_str = row['Source'].replace(" ", "_")  # Replace spaces with underscores for filename
    filename = f"{date_str}_{source_str}_{index}.txt"
    
    # Define the file path
    file_path = os.path.join(output_folder, filename)
    
    # Ensure the content is a string (handle NaN or non-string types)
    content = str(row['cleaned_corpus']) if pd.notna(row['cleaned_corpus']) else ""

    # Write the content of "cleaned_corpus" to the file
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(content)

    print(f"Created file: {filename}")
    
    # Increment the file counter
    file_count += 1

# Print the total number of files created
print(f"\nTotal number of .txt files created: {file_count}")
```
